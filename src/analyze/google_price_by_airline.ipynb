#### 전체 항공사의 인천공항 점유율 확인하기

```python
import pandas as pd
import matplotlib.pyplot as plt

wedge_props = {"width" : 0.7, "edgecolor" : "w", "linewidth" : 3}
# 파이 차트 그리기
plt.figure(figsize=(15, 15))
df_pie = airline_in_incheon_01_12_af_processing.sort_values(by="출발", ascending=False)

plt.pie(df_pie["출발"], labels=df_pie['항공사'], autopct='%1.1f%%', startangle=0, wedgeprops=wedge_props)
plt.title('인천공항 점유율')
# plt.legend(loc="upper left", bbox_to_anchor=(0.85,1))

plt.show()
```
